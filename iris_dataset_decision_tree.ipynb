## Creating a Data model using decision trees

```python
from sklearn.tree import DecisionTreeClassifier as dtc
```

```python
# Defining and fitting the DecisionTreeClassifier instance
tree1 = dtc()

# passing the features, labels
tree1.fit(x, y)
```

```python
# Texual representation of this decision tree
from sklearn import tree

text_representation = tree.export_text(tree1)
print(text_representation)
```

```python
# Classical representation of this decision tree
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25, 20))
_ = tree.plot_tree(tree1,
                   feature_names=iris.feature_names,
                   class_names=iris.target_names,
                   filled=True)
```

```python
# Plot Decision Tree with dtreeviz Package
# !pip install dtreeviz
from dtreeviz.trees import dtreeviz # remember to load the package

viz = dtreeviz(tree1, x, y,
                target_name="target",
                feature_names=iris.feature_names,
                class_names=list(iris.target_names))

viz
```

## Prediction

```python
sample_one_pred_tree1 = int(tree1.predict([[5, 5, 1, 3]]))
iris.target_names[sample_one_pred_tree1]
```
